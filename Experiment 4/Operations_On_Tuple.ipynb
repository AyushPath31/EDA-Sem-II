```python
# Python Tuples
# Access tuple items
cars = ("Ferrari","McLaren","Mercedes","Aston Martin","Honda","Audi")
print(cars[2])
```

```python
# Tuple is immutable
brands = ("Gucci","Lifestyle","Rolex","Casio","Titan","Sonata")
brands[2] = "Boat"
print(brands)
```

```python
# Addition of Tuple
t1 = (1,3,7,9)
t2 = (2,4,6,8)
t3 = t1 + t2
print("t1: ",t1)
print("t2: ",t2)
print("t3: ",t3)
```

```python
# Address Verification after appending new item in the tuple
cars = ("Ferrari","McLaren","Mercedes","Aston Martin","Honda","Audi")
print("Before ID: ",id(cars))
new = ("RedBull",)
cars += new
print("After: ",id(cars))
```

```python
# Address Verification after appending new item in the list
cars = ["Ferrari","McLaren","Mercedes","Aston Martin","Honda","Audi"]
print("Before ID: ",id(cars))
new = ("RedBull")
cars += new
print("After: ",id(cars))
```

```python
# Problem Statement 2 - Employee Attendance
# Store Attendance in tuple; Count and display total no. of days present and absent
# Check if employee absent atleast once

record = ("P","P","A","P","A","A","P","P","P","P")
print("Total Days Present: ",record.count("P"))
print("Total Days Absent:  ",record.count("A"))

absent = record.count("A")
if absent > 1:
  print("You have been absent more than once. Absent days = ",absent)
else:
  print("100% Attendance")

```

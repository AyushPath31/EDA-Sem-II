```python
# Problem Statement 1 - Find Max Min and Avg Marks from a list

marks = [78,85,90,66,88]
print("Maximum marks: ",max(marks))
print("Minimum marks: ",min(marks))
print("Average marks: ",sum(marks) / len(marks),end='\n\n')

marks.sort(reverse=False)
print("Marks in Ascending Order :",marks)
marks.sort(reverse=True)
print("Marks in Descending Order :",marks)
```

```python
# Problem Statement 2 -
# Create Grocery List
# Add New items to this list
# Remove purchased items
# Display the final shopping list

market = ["Lettuce","Bread","Cheese","Mayo","Sauce","Patty"]

market.append("Butter") # Butter is added at the end
print("After Append: ",market,end='\n\n')

market.remove("Mayo") # Mayo is removed from list
print("After Remove: ",market,end='\n\n')

market.insert(1,"Water Bag") # Water bag is added at index 1 after Lettuce
print("After Insert",market,end='\n\n')

market.pop(4) # Sauce is at index 4 so it is removed
print("After Pop Obj: ",market,end='\n\n')
```

```python
# Problem Statement 3 -
# Create a list of contacts; Add New contact; Delete an existing contact, Display Contacts alphabetically

contacts = ["Jesus","Buddha","Vishnu","Sun Tzu","Pheonix","Fishman"]
print("Before Insert: ",contacts,end='\n\n')

contacts.insert(2,"Monkey King")
print("After Insert: ",contacts,end='\n\n')
contacts.remove("Fishman")

contacts.sort(reverse=False)
print("Alphabetical Order: ",contacts,end='\n\n')
```

```python
# Problem Statement 4 -
# Daily Temp readings of a city are stored in a list for 10 days
# Display Temp of first and last 5 Days; Find Highest and Lowest Temp; Calc Avg Temp

temp = [13,19,21,25,12,22,27,23,33,11]
print("Temp for 10 Days: ",temp, end='\n\n')
print("Temp of First 5 Days: ",temp[5:])
print("Temp of Last 5 Days: ",temp[-5:],end='\n\n')

print("Max Temp: ",max(temp))
print("Min Temp: ",min(temp),end='\n\n')

print("Avg Temp: ",sum(temp)/len(temp))
```

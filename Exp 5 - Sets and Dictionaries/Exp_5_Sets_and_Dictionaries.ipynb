Sets are used to store multiple items in a single variable  
Sets items are ordered, unchangeable, and do not allow duplicate values. Unordered means that items in a set do not have a defined order. Set items can appear in a different order everytime you use them.


```python
# Creation of set
myset = {"Ferrari","Haas","McLaren","Mercedes","Audi","Aston Martin"}
print(myset)
```

```python
# Duplicates not allowed
myset = {"Ferrari","Haas","McLaren","Ferrari","Ferrari"}
print(myset)
```

```python
# The values True and 1 are considered the same values in sets
myset = {"Ferrari","Haas","McLaren","Mercedes","Audi","Aston Martin",True, 1, 2, 3, 4}
print(myset)
```

```python
# A set with strings, integers and boolean values:
setA = {"xyz",99,True,False,44,"wild west",78.992}
setA
```

```python
# To check membership
myset = {"Ferrari","Haas","McLaren","Ferrari","Ferrari"}
print("BMW" in myset)
print("Haas" in myset)
```

```python
# Once a set is created, you cannot replace its items, but you can add new items.
myset = {"Ferrari","Haas","McLaren","Mercedes","Audi","Aston Martin"}
myset.add("BMW")
print(myset)
```

```python
# Remove Items
myset = {"Ferrari","Haas","McLaren","Mercedes","Audi","Aston Martin"}
myset.remove("Audi")
print(myset)
```

```python
# Set Operations
A = {1,2,3,4}
B = {3,4,5,6}
print("Union : ",A | B)
print("Intersection : ",A & B)
print("Difference : ",A - B)
print("Symmetric Difference", A ^ B)
```

## Python FrozenSet  
frozenset is a immutable version of set.

```python
# frozenset
myset = frozenset({"Ferrari","Haas","McLaren","Mercedes","Audi","Aston Martin"})
print(myset)
print(type(myset))
```

```python
# Adding or removal of elements is not allowed in frozenset
myset = frozenset({"Ferrari","Haas","McLaren","Mercedes","Audi","Aston Martin"})
myset.add("BMW")
myset.remove("Audi")
print(myset)

```

```python
# Merge frozensets using union
myset = frozenset({"Ferrari","McLaren","Mercedes","Aston Martin"})
newset = frozenset({"Apple","BlackBerry","Cherry"})
print("Union : ", myset | newset)
```

```python
'''Problem Statement 1 : Unique Event Participants
A college even has a list of students who registered multiple times by mistake.
Write a program using set to store participant names and display only unique participants'''

registered_names = ["Sam Altman","Elon Musk","Satya Nadella","Jeff Bezos","Sundar Pichai","Marc Benioff","Elon Musk","Sam Altman","Jeff Bezos"]
unique_names = set(registered_names)
print("Unique Names: ",unique_names)
```

```python
'''Problem Statement 2 : Three students choose elective subjects.
Find the common subjects chosen by both students using set operations.'''

student1_electives = {"Software","Hardware","Cloud","CyberSec","Semicon","AIML","Mechanical","Designing"}
student2_electives = {"Cloud","Software","Painting","Sleeping","Running","Flying"}
student3_electives = {"Swimming","Laughing","Running","Cloud","Software"}
print("Common Sub :", student1_electives & student2_electives & student3_electives)
```

```python
'''Problem Statement 3 : One set contains students in the Cricket Club, another in Football club.
Find students who are :
1. In both clubs & 2. In only one club'''

cricket_club = {"Alex","Ryan","Austin","Tyler","Rob","Mark","Joe"}
football_club = {"Ryan","Mark","Joe","Bruce","Tony","Austin"}
print("Students in Both Clubs: ",cricket_club & football_club)
print("Student in either one club: ",cricket_club ^ football_club)
```

```python
'''Problem Statement 4 : Absent Students
Given a set of all students and set of present students, find the absent students.'''
all_students = {"monkey man","kangaroo man","spider man","ant man","zebra man","owl man","dolphin man","boring man"}
present_students = {"spider man","ant man","owl man","zebra man"}
print("Absent Students : ", all_students-present_students)
```

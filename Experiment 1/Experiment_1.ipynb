```python
# Question 1 : Take marks of 5 subjects. Find sum and average

print("Enter marks for your subjects \n")
eng = int(input("English : "))
math = int(input("Math : "))
phys = int(input("Physics : "))
chem = int(input("Chemistry : "))
bio = int(input("Biology : "))
total = eng+math+phys+chem+bio

print("Your total marks = ",total)
print("Your average marks = ",total/5)
```

```python
# Question 3 :Find area of a rectangle

print("To find Area of a Rectangle -")
l = int(input("Enter Value of length = "))
b = int(input("Enter Value of breadth = "))

print("Area of Rectangle = ",l*b)
```

```python
# Question 4 : Find perimeter of rectangle

print("To find Perimeter of Rectangle -")
l1 = int(input("Enter Value of length = "))
b1 = int(input("Enter Value of breadth = "))

print("Perimeter of Rectangle = ",l1*2+b1*2)
```

```python
# Question 5 : Simple Calculator

print("Simple Calculator")
print(" Operations : \n 1. Add \n 2. Subtract \n 3. Multiply \n 4. Divide \n")
choice = int(input("Choose Your Operation Number : "))
num1 = int(input("Enter 1st Number : "))
num2 = int(input("Enter 2nd Number : "))

match choice:

 case 1:
  print("Add : %d + %d = %d" % (num1,num2,num1+num2))

 case 2:
  print("Subtract : %d - %d = %d" % (num1,num2,num1+num2))

 case 3:
  print("Multiply : %d x %d = %d" % (num1,num2,num1*num2))

 case 4:
  print("Divide : %d / %d = %d" % (num1,num2,num1/num2))

 case _:
  print("Invalid Choice" )
```
